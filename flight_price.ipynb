#### EDA ANd Feature Engineering Flight Price Prediction

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel('flight_price.xlsx')
df.head()
```

```python
## get the basic info about the data
df.info()
```

```python
df.describe()
```

```python
##Feature Engineering
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df[['Date','Month','Year']]
```

```python
df.head()
```

```python
df.info()
```

```python
## Coverting Object type to df
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
## Drop Date of Journey  Column
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['Arrival_hour']=df['Arrival_Time'].str.split(':').str[0].astype(int)
df['Arrival_minute']=df['Arrival_Time'].str.split(':').str[1].astype(int)
```

```python
df[['Arrival_hour','Arrival_minute']].info()
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Dep_hour']=df['Dep_Time'].str.split(':').str[0].astype(int)
df['Dep_min']= df['Dep_Time'].str.split(':').str[1].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Duration_hour']=df['Duration'].str.split(' ').str[0]
df['Duration_min']= df['Duration'].str.split(' ').str[1]
```

```python
df.head()
```

```python
df.drop('Duration',axis=1,inplace=True)
```

```python
df['Total_Stops'].unique()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df['Total_Stops'].mode()
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, np.nan:1, '4 stops':4})
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df.head()
```

```python
df.drop('Route',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df['Duration_hour']=df['Duration_hour'].str.split('h').str[0]
df['Duration_min']=df['Duration_min'].str.split('m').str[0]
```

```python
mode_value=df[df['Duration_min'].notna()]['Duration_min'].mode()[0]
```
